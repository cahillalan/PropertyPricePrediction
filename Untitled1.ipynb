## Wilcox Test

"The Wilcoxon signed-rank test is the nonparametric test equivalent to the dependent t-test." This means that the wilcox test does not assume or depend on the data being normally distributed. The wilcox test will allow us to see if there is a difference between the first parameters and the second parameters. 

In this case we will use the wilcox test to determine if the price of a house changes based off of it type. It is not possible to check the distribution off hand so an array of wilcox tests and anova tests will be utilised.

Below is the failed attempt at shapiro normality test

```r
shapiro.test(all$price)
```

Below is the first attempt at wilcox.test. As can be viewed the outputs provide two values a W and a p-value. We will not be utilising the W value but instead concentrate on the p-value. The p-vale will determine if we can accept or reject the null hypothesis. In each case the hypothesis is that the value of the price will change dramatically based upon the change in the second parameter.

For this to be true we must accept the null hypothesis.

A p-value of below .05 means we can accept the null hypothesis.

As can be viewed below the p-value is 2.2e-16. This equates to a value significantly lower than 0.05, meaning our hypothesis can be considered true.

```r
wilcox.test(all$price[all$type == "Detached House "] , all$price[all$type == "Terraced House "])
```

Amazingly the hypothesis for the second test is rejected as false. The reason for this is the p-value is above .05 with a value of .06322

```r
wilcox.test(all$price[all$type == "Detached House "] , all$price[all$type == "New Dwelling House/Apartment"])
```
